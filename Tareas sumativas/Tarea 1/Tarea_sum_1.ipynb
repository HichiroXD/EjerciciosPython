```python
# Visiualizamos la distribución de la variable 'Pclass' en el dataset de entrenamiento
sns.countplot(x='Pclass', data=train_data) # Crea un gráfico de barras para mostrar la distribución de la variable 'Pclass' en el DataFrame train_data
plt.title('Distribución de Clases') # Agrega un título al gráfico
plt.xlabel('Clase') # Agrega una etiqueta al eje x
plt.ylabel('Cantidad') # Agrega una etiqueta al eje y
plt.show() # Muestra el gráfico en pantalla
```

```python
# Ver posibilidad de superviviencia por clase
sns.countplot(x='Pclass', hue='Survived', data=train_data) # Crea un gráfico de barras para mostrar la distribución de la variable 'Pclass' con respecto a la variable 'Survived' en el DataFrame train_data
plt.title('Supervivencia por Clase') # Agrega un título al gráfico
plt.xlabel('Clase') # Agrega una etiqueta al eje x
plt.ylabel('Cantidad') # Agrega una etiqueta al eje y
plt.legend(title='Supervivió', labels=['No', 'Sí']) # Agrega una leyenda al gráfico para indicar qué color representa a los sobrevivientes y no sobrevivientes
plt.show() # Muestra el gráfico en pantalla
```

```python
# Ver posibilidad de superviviencia por sexo
sns.countplot(x='Sex', hue='Survived', data=train_data) # Crea un gráfico de barras para mostrar la distribución de la variable 'Sex' con respecto a la variable 'Survived' en el DataFrame train_data
plt.title('Supervivencia por Sexo') # Agrega un título al gráfico
plt.xlabel('Sexo') # Agrega una etiqueta al eje x
plt.ylabel('Cantidad') # Agrega una etiqueta al eje y
plt.legend(title='Supervivió', labels=['No', 'Sí']) # Agrega una leyenda al gráfico para indicar qué color representa a los sobrevivientes y no sobrevivientes
plt.show() # Muestra el gráfico en pantalla
```

```python
# Grafico de supervivencia por clase
survival_rate_by_class = train_data.groupby('Pclass')['Survived'].mean() # Agrupa el DataFrame train_data por la columna 'Pclass' y calcula la media de la columna 'Survived' para cada clase, lo que representa el porcentaje de supervivencia por clase
print(survival_rate_by_class) # Imprime el porcentaje de supervivencia por clase
```

```python
# Porcertaje de supervivencia por sexo
survival_rate_by_sex = train_data.groupby('Sex')['Survived'].mean() # Agrupa el DataFrame train_data por la columna 'Sex' y calcula la media de la columna 'Survived' para cada sexo, lo que representa el porcentaje de supervivencia por sexo
print(survival_rate_by_sex) # Imprime el porcentaje de supervivencia por sexo
```

```python
# grafico de supervivencia por edad
sns.histplot(data=train_data, x='Age', hue='Survived', multiple='stack') # Crea un histograma de la variable 'Age' con respecto a la variable 'Survived' en el DataFrame train_data, apilando las barras para mostrar la distribución de edades entre sobrevivientes y no sobrevivientes
plt.title('Supervivencia por Edad') # Agrega un título al gráfico
plt.xlabel('Edad') # Agrega una etiqueta al eje x
plt.ylabel('Cantidad') # Agrega una etiqueta al eje y
plt.legend(title='Supervivió', labels=['No', 'Sí']) # Agrega una leyenda al gráfico para indicar qué color representa a los sobrevivientes y no sobrevivientes
plt.show() # Muestra el gráfico en pantalla
```

```python
# Porcertaje de supervivencia por SibSp
survival_rate_by_sibsp = train_data.groupby('SibSp')['Survived'].mean() # Agrupa el DataFrame train_data por la columna 'SibSp' y calcula la media de la columna 'Survived' para cada número de hermanos/cónyuges a bordo, lo que representa el porcentaje de supervivencia por número de hermanos/cónyuges
print(survival_rate_by_sibsp) # Imprime el porcentaje de supervivencia por número de hermanos/cónyuges a bordo
```

```python
# Porcentaje de supervivencia por parch
survival_rate_by_parch = train_data.groupby('Parch')['Survived'].mean() # Agrupa el DataFrame train_data por la columna 'Parch' y calcula la media de la columna 'Survived' para cada número de padres/hijos a bordo, lo que representa el porcentaje de supervivencia por número de padres/hijos
print(survival_rate_by_parch) # Imprime el porcentaje de supervivencia por número de padres/hijos a bordo
```

```python
# Porcentaje de supervivencia por clase y sexo 
survival_rate_by_class_and_sex = train_data.groupby(['Pclass', 'Sex'])['Survived'].mean() # Agrupa el DataFrame train_data por las columnas 'Pclass' y 'Sex', y calcula la media de la columna 'Survived' para cada combinación de clase y sexo, lo que representa el porcentaje de supervivencia por clase y sexo
print(survival_rate_by_class_and_sex) # Imprime el porcentaje de supervivencia por clase y sexo
```

```python
# Tasa de supervivencia de niños (menores de 18 años)
children = train_data[train_data['Age'] < 18]
survival_rate_children = children['Survived'].mean()
print(f"Tasa de supervivencia de niños (menores de 18 años): {survival_rate_children:.2%}")
print(f"Total de niños en el dataset: {len(children)}")
print(f"Niños que sobrevivieron: {children['Survived'].sum()}")
```

```python
# Visualización: Supervivencia de Niños por Clase de Pasaje
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico de barras con conteo
children.groupby('Pclass')['Survived'].value_counts().unstack().plot(kind='bar', ax=axes[0])
axes[0].set_title('Supervivencia de Niños por Clase de Pasaje (Conteo)')
axes[0].set_xlabel('Clase de Pasaje')
axes[0].set_ylabel('Cantidad')
axes[0].legend(['No Sobrevivió', 'Sobrevivió'], title='Supervivió')
axes[0].set_xticklabels(['Clase 1', 'Clase 2', 'Clase 3'], rotation=0)

# Gráfico de porcentage
children.groupby('Pclass')['Survived'].mean().plot(kind='bar', ax=axes[1], color='coral', edgecolor='black')
axes[1].set_title('Tasa de Supervivencia de Niños por Clase')
axes[1].set_xlabel('Clase de Pasaje')
axes[1].set_ylabel('Tasa de Supervivencia')
axes[1].set_xticklabels(['Clase 1', 'Clase 2', 'Clase 3'], rotation=0)
axes[1].set_ylim([0, 1])

plt.tight_layout()
plt.show()
```

```python
# Visualización: Supervivencia de Niños por Género
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico de barras con conteo
children.groupby('Sex')['Survived'].value_counts().unstack().plot(kind='bar', ax=axes[0])
axes[0].set_title('Supervivencia de Niños por Género (Conteo)')
axes[0].set_xlabel('Género')
axes[0].set_ylabel('Cantidad')
axes[0].legend(['No Sobrevivió', 'Sobrevivió'], title='Supervivió')
axes[0].set_xticklabels(['Femenino', 'Masculino'], rotation=0)

# Gráfico de porcentage
children.groupby('Sex')['Survived'].mean().plot(kind='bar', ax=axes[1], color=['#FF69B4', '#4169E1'], edgecolor='black')
axes[1].set_title('Tasa de Supervivencia de Niños por Género')
axes[1].set_xlabel('Género')
axes[1].set_ylabel('Tasa de Supervivencia')
axes[1].set_xticklabels(['Femenino', 'Masculino'], rotation=0)
axes[1].set_ylim([0, 1])

plt.tight_layout()
plt.show()
```

```python
# Visualización: Supervivencia de Niños vs Adultos
fig, axes = plt.subplots(1, 2, figsize=(50, 10))

# Gráfico de barras con conteo
survival_by_age_group = train_data.copy()
survival_by_age_group.groupby('Age')['Survived'].value_counts().unstack().plot(kind='bar', ax=axes[0])
axes[0].set_title('Supervivencia de Niños vs Adultos (Conteo)')
axes[0].set_xlabel('Grupo de Edad')
axes[0].set_ylabel('Cantidad')
axes[0].legend(['No Sobrevivió', 'Sobrevivió'], title='Supervivió')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)

# Gráfico de porcentage
survival_by_age_group.groupby('Age')['Survived'].mean().plot(kind='bar', ax=axes[1], color='skyblue', edgecolor='black')
axes[1].set_title('Tasa de Supervivencia: Niños vs Adultos')
axes[1].set_xlabel('Grupo de Edad')
axes[1].set_ylabel('Tasa de Supervivencia')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=0)
axes[1].set_ylim([0, 1])

plt.tight_layout()
plt.show()
```

```python
# Supervivencia de niños por clase de pasaje
children_by_class = children.groupby('Pclass')['Survived'].agg(['count', 'sum', 'mean'])
children_by_class.columns = ['Total', 'Sobrevivieron', 'Tasa de Supervivencia']
print("\nSupervivencia de Niños por Clase de Pasaje:")
print(children_by_class)
```

```python
# Supervivencia de niños por género
children_by_sex = children.groupby('Sex')['Survived'].agg(['count', 'sum', 'mean'])
children_by_sex.columns = ['Total', 'Sobrevivieron', 'Tasa de Supervivencia']
print("\nSupervivencia de Niños por Género:")
print(children_by_sex)
```

```python
# Comparación de supervivencia: Niños vs Adultos
survival_by_age_group = train_data.copy()
survival_by_age_group['Grupo de Edad'] = survival_by_age_group['Age'].apply(lambda x: 'Niño' if x < 18 else 'Adulto')
comparison = survival_by_age_group.groupby('Grupo de Edad')['Survived'].agg(['count', 'sum', 'mean'])
comparison.columns = ['Total', 'Sobrevivieron', 'Tasa de Supervivencia']
print("\nComparación de Supervivencia por Grupo de Edad:")
print(comparison)
```

```python
# ANÁLISIS INTEGRAL DE SUPERVIVENCIA DE NIÑOS EN EL TITANIC

# 1. Resumen general
print(f"   • Tasa de supervivencia de niños: {survival_rate_children:.2%}")
print(f"   • Total de niños analizados: {len(children)}")
print(f"   • Niños que sobrevivieron: {int(children['Survived'].sum())}")
print(f"   • Niños que murieron: {len(children) - int(children['Survived'].sum())}")

# 2. Comparativa niños vs adultos
survival_niños = survival_rate_children

# Calcular supervivencia de adultos
adultos = train_data[train_data['Age'] >= 18]
survival_adultos = adultos['Survived'].mean()
diferencia = survival_niños - survival_adultos
print(f"   • Supervivencia de niños: {survival_niños:.2%}")
print(f"   • Supervivencia de adultos: {survival_adultos:.2%}")
print(f"   • Diferencia: {diferencia:+.2%} {'(niños favorecidos)' if diferencia > 0 else '(adultos favorecidos)'}")

# 3. Análisis por género
niñas = children[children['Sex'] == 'female']['Survived'].mean()
niños = children[children['Sex'] == 'male']['Survived'].mean()
print(f"   • Supervivencia de niñas: {niñas:.2%}")
print(f"   • Supervivencia de niños: {niños:.2%}")
print(f"   • Diferencia: {(niñas - niños):.2%}")

# 4. Análisis por clase de pasaje
for clase in [1, 2, 3]:
    survival_clase = children[children['Pclass'] == clase]['Survived'].mean()
    total_clase = len(children[children['Pclass'] == clase])
    print(f"   • Clase {clase}: {survival_clase:.2%} ({int(children[children['Pclass'] == clase]['Survived'].sum())}/{total_clase})")

# 5. Niñas en clase 1
niñas_clase1 = children[(children['Sex'] == 'female') & (children['Pclass'] == 1)]['Survived'].mean()
count_niñas_clase1 = len(children[(children['Sex'] == 'female') & (children['Pclass'] == 1)])
survived_niñas_clase1 = int(children[(children['Sex'] == 'female') & (children['Pclass'] == 1)]['Survived'].sum())
print(f"   • Supervivencia: {niñas_clase1:.2%} ({survived_niñas_clase1}/{count_niñas_clase1})")

print(""""
      CONCLUSIONES PRINCIPALES:
      """)
print("""
1. POLÍTICA "MUJERES Y NIÑOS PRIMERO":
   Los niños tenían una alta tasa de supervivencia, especialmente en
   clases altas (probabilidad de evacuación preferente).

2. FACTOR CLASE SOCIOECONÓMICA:
   La clase de pasaje fue determinante. Los niños de 1ª clase tuvieron
   mejores oportunidades que los de 3ª clase.

3. FACTOR GÉNERO ENTRE NIÑOS:
   La preferencia de "mujeres y niños primero" se reflejó incluso entre
   los menores, con niñas teniendo tasas más altas.

4. DISPARIDAD SOCIAL:
   Los niños de clase baja enfrentaron tasas de supervivencia
   significativamente más bajas, mostrando desigualdad incluso en
   situaciones de emergencia.
""")
```

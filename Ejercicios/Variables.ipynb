# Tipos de variables en python

## Variables simples:
 * int: Números enteros.
 * srt:Cadena de caracteres.
 * float: Números decimales.
 * bool: Booleanos (True o False).

 ## Variables compuestas:
 * list: [] Datos mutables, se puede agregar, quitar y modificar elementos.
 * tuplas: () Datos inmutables, **no** se puede agregar ni quitar elementos. Hay orden.
 * set: {} Datos no repetidos, no hay orden.
 * dict: {key : value} Datos tipo set.

```python
a = 3
b = 4.0
c = "Cadena"
d = False
print(str(a) + " es un número entero")
print(str(b) + " es un número flotante")
print(c + " es una cadena de texto")
print(str(d) + " es un valor booleano")
```

### Operaciones Simples Compuestas

#### Listas

```python
l = [1, 2, 3, 4, 5] # Lista de números del 1 al 5
l[0] # Accede al primer elemento de la lista (1)
l[-1] # Accede al último elemento de la lista (5)
l[1:4] # Accede a los elementos desde el índice 1 hasta el índice 3 (2, 3, 4)
print(l[0]) # Imprime el primer elemento de la lista (1)
print(l[-1]) # Imprime el último elemento de la lista (5)
print(l[1:4]) # Imprime los elementos desde el índice 1 hasta el índice 3 (2, 3, 4)
print(l) # Imprime la lista completa
```

```python
l_1 = [1, 2.0, "tres", True] # Lista con diferentes tipos de datos
print(l_1) # Imprime la lista completa
```

```python
l_1.append(4) # Agrega el número 4 al final de la lista
print(l_1) # Imprime la lista actualizada con el nuevo elemento
```

```python
l_1.pop() # Elimina el último elemento de la lista (4)
print(l_1) # Imprime la lista actualizada después de eliminar el último elemento
```

#### Tuplas

```python
t = (1, 2.0, "tres", True) # Tupla con diferentes tipos de datos
print(t) # Imprime la tupla completa
```

```python
t_1 = (1,2,3,4) # Tupla de números del 1 al 4
print(t_1) # Imprime la tupla completa
```

```python
t_2 = (1,1,1,1,1,2,3,4) # Tupla con números repetidos
print(t_2) # Imprime la tupla completa con números repetidos
```

```python
t_2.count(1) # Cuenta cuántas veces aparece el número 1 en la tupla t_2
t_2.index(2) # Devuelve el índice de la primera aparición del número 2 en la tupla t_2
print("El numero " + str(1) + " aparece " + str(t_2.count(1)) + " veces en la tupla t_2") # Imprime un mensaje con el número de veces que aparece el número 1 en la tupla t_2
print("El numero " + str(2) + " aparece en la posición " + str(t_2.index(2)) + " en la tupla t_2") # Imprime un mensaje con el índice de la primera aparición del número 2 en la tupla t_2
```

#### Diccionarios

```python
dic = {"clave1": "valor1", "clave2": 2, "clave3": [1, 2, 3]} # Diccionario con diferentes tipos de datos
dic_1 = {"nombre": "Juan", "edad": 30, "hobbies": ["futbol", "lectura"]} # Diccionario multinivel con diferentes tipos de datos
print(dic_1) # Imprime el diccionario completo
```

```python
dic_1["nombre"] # Accede al valor asociado a la clave "nombre" en el diccionario dic_1 (Juan)
dic_1["edad"] # Accede al valor asociado a la clave "edad" en el diccionario dic_1 (30)
dic_1["hobbies"] # Accede al valor asociado a la clave "hobbies" en el diccionario dic_1 (["futbol", "lectura"])
print(dic_1["nombre"]) # Imprime el valor asociado a la clave "nombre" en el diccionario dic_1 (Juan)
print(dic_1["edad"]) # Imprime el valor asociado a la clave "edad"
print(dic_1["hobbies"]) # Imprime el valor asociado a la clave "hobbies" en el diccionario dic_1 (["futbol", "lectura"])
```

#### Condicionales

```python
if dic_1["hobbies"] is None:
    print("El usuario no tiene hobbies")
else:
    print("El usuario tiene hobbies: " + ", ".join(dic_1["hobbies"])) # Imprime un mensaje con los hobbies del usuario, separados por comas
```

```python
if a > b: # Compara si a es mayor que b
    print(str(a) + " es mayor que " + str(b))
elif a < b: # Compara si a es menor que b
    print(str(a) + " es menor que " + str(b))
else: # Si a no es ni mayor ni menor que b, entonces son iguales
    print(str(a) + " es igual a " + str(b))
```

#### Excepciones

```python
try:
    resultado = a / 0 # Intenta dividir a entre 0, lo cual generará un error
except ZeroDivisionError: # Captura el error de división por cero
    print("Error: No se puede dividir por cero") # Imprime un mensaje de error
```

```python
try:
    dic_1["direccion"] # Intenta acceder a una clave que no existe en el diccionario dic_1 
except KeyError: # Captura el error de clave no encontrada
    print("Error: La clave 'direccion' no existe en el diccionario dic_1") # Imprime un mensaje de error indicando que la clave no existe   
```

#### Loop

```python
for i in range(5): # Itera desde 0 hasta 4
    print(i) # Imprime el valor actual de i en cada iteración
```

```python
# Compresion de listas
cuadrados = [x**2 for x in range(10)] # Crea una lista de los cuadrados de los números del 0 al 9 utilizando comprensión de listas
print(cuadrados) # Imprime la lista de cuadrados
```

```python
l_c = [l**2 for l in range(5)] # Crea una lista de los números del 0 al 4 elevados al cuadrado utilizando comprensión de listas
print(l_c) # Imprime la lista de números elevados al cuadrado
```

```python
l_c_2 =[l/2 for l in range(5) if l != 0] # Crea una lista de los números del 0 al 4 divididos entre 2 utilizando comprensión de listas
print(l_c_2) # Imprime la lista de números divididos entre 2 y viendo si la condicion se cumple.
```

```python
# Uso de tuplas en compresion de listas
t_c = [(x, x**2) for x in range(5)] # Crea una lista de tuplas donde cada tupla contiene un número y su cuadrado utilizando comprensión de listas
print(t_c) # Imprime la lista de tuplas con números y sus cuadrados
```

```python
# Uso de while
contador = 0 # Inicializa un contador en 0
while contador < 5: # Continúa iterando mientras el contador sea menor que 5
    print(contador) # Imprime el valor actual del contador en cada iteración
    contador += 1 # Incrementa el contador en 1 en cada iteración
```
